```python
#feature_1
weather = ['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny',
           'Sunny','Rainy','Sunny','Overcast','Overcast','Rainy']
```

```python
#feature_2
temp = ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild'
       ,'Hot','Mild']
```

```python
#label
play = ['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes',
       'No']
print(len(play))
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
weather_encoded = le.fit_transform(weather)
weather_encoded
```

```python
temp_encoded = le.fit_transform(temp)


temp_encoded
```

```python
label_encoded = le.fit_transform(play)
label_encoded
```

```python
features = list(zip(weather_encoded,temp_encoded))
features
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 3)#n_neighbors=sqrt(sum(f1)) 
```

```python
model.fit(features,label_encoded)
```

```python
#weather
# sunny=2
# overcast=0
# rainy=1

#temp
# Hot= 1
# Mild=2
# cold=0 

lst = []

for i in range(0, 2):
    ele = int(input())
    lst.append(ele) 
predicted = model.predict([lst])
if(predicted== [0]):
    print("No")
else:
    print("Yes")

# print(predicted)



 
```
